## Uma **função** é uma sequência de comandos definida com um nome via **def**. Ela pode ter parâmetros e retornar um valor via **return** ou **yield**. Somente é executada quando chamada. **return** e **yield** não são funções, apenas palavras reservadas. Também existem **lambda** funções, mais avançadas.

```
def f(a):
  a = a + 5
  return a

b = 0
f(b)
print(b, ', ', end= ' ')
b = f(b)
print(b)
```

```
def f(x):
  print('x', end= ' ')
  if x <= 1:
    return 1
  else:
    return x + f(x-1)
```

```
f(4)
```

```
def comum(seq1, seq2):
  res = []
  for x in seq1:
    if x in seq2:
      res.append(x)
  return res
```

```
comum('azul', 'amarelo')
```

```
comum(range(5), [1, 3, 5])
```

```
comum('azul', ['a', 'b'])
```

Variáveis globais não são alteradas dentro de funções, a menos que declaradas como global dentro delas.

```
a = 'X'
def func():
  a = 'O'

func()
print (a)
```

```
a = 'X'
def func():
  global a
  a = 'O'

func()
print (a)
```

yield é um gerador, podemos utilizá-lo em uma função onde cada elemento é gerado online via **next()**

```
def fat():
  n = 1
  f = 1
  while True:
    f = f * n
    yield f
    n = n + 1

a = fat()
for i in range(5):
    print(next(a), end= ' ')
```

```
def fib():
  a, b = 1, 1
  while True:
    yield a
    a, b = b, a + b

a = fib()
for i in range(5):
    print(next(a), end= ' ')
```
